```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import plotly.express as px # plotly express
%matplotlib inline
import os

# Input data files are available in the '/kaggle/input' or '../../../datasets/extracts/' directory.
file_input=['/kaggle/input','../../../datasets/']
files={}
for dirname, _, filenames in os.walk(file_input[1]):
    for filename in filenames:
        if 'csv' in filename:
            files[filename.replace('.csv','')]=os.path.join(dirname, filename)
            print(filename.replace('.csv',''))
```

```python
province_df=pd.DataFrame()

df= pd.read_csv(files['covid_19_data'])
df=df[df['Province/State'].isin(['Reunion','Hong Kong','Quebec','New York','Virginia'])][['ObservationDate','Province/State','Confirmed','Deaths','Recovered']]\
    .rename({'ObservationDate':'Date','Province/State':'Province'},axis=1)
df['Date']= pd.to_datetime(df['Date'],format='%m/%d/%Y').dt.strftime('%Y-%m-%d')
province_df= pd.concat([province_df,df])
```

```python
df= pd.read_csv(files['covid_19_india'])
df= df[df['State/UnionTerritory']=='Maharashtra'][['Date','State/UnionTerritory','Confirmed','Deaths','Cured']]\
    .rename({'State/UnionTerritory':'Province','Cured':'Recovered'},axis=1)
df['Date']= pd.to_datetime(df['Date'],format='%d/%m/%y').dt.strftime('%Y-%m-%d')
province_df= pd.concat([province_df,df])
```

```python
df= pd.read_csv(files['covid19_italy_region'])
df= df[df['RegionName']=='Lombardia'][['Date','RegionName','TotalPositiveCases','Deaths','Recovered']]\
    .rename({'RegionName':'Province','TotalPositiveCases':'Confirmed'},axis=1)
df['Date']=pd.to_datetime(df['Date']).dt.strftime('%Y-%m-%d')
province_df= pd.concat([province_df,df])
```

```python
weather_keys={'Reunion':'Weather Reunion-France',
              'Hong Kong':'Weather Hong Kong',
              'New York':'Weather New-York',
              'Maharashtra':'Weather Maharashtra',
              'Lombardia':'Weather Lombardia-Italy',
              'Virginia':'Weather Virginia',
              'Quebec':'Weather Quebec',
             }
pd.set_option('display.max_columns', 500)

total_weather_df=pd.DataFrame()
for key in weather_keys.items():
    weather_df= pd.read_csv(files[key[1]])
    weather_df['Date']=(pd.to_datetime(weather_df['valid_time_gmt'],unit='s') - pd.DateOffset(14)).dt.strftime('%Y-%m-%d')
    weather_df= weather_df[['Date','temp','dewPt','wspd','pressure','pressure_tend','heat_index','rh','vis','wc','wdir','feels_like','uv_index']].groupby(['Date']).agg(['min','mean','max']).reset_index()
    weather_df.columns = weather_df.columns.map('| '.join).str.strip('| ')
    weather_df['Province']=key[0]
    
    total_weather_df=pd.concat([total_weather_df,weather_df])
    
province_weather_df= pd.merge(province_df,total_weather_df,left_on=['Date','Province'],right_on=['Date','Province'],how='left')

```

```python
province_weather_df= province_weather_df[~province_weather_df['temp| min'].isna()]
```

```python

tmp=province_weather_df[province_weather_df['Province']=='Hong Kong']

fig=px.line(province_weather_df,
       x='Date',
       y='temp| min',
       color='Province'
       )
fig.update_layout(yaxis=dict(type='linear'))
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,10)
corr=province_weather_df[['Confirmed','Recovered','Deaths','temp| min','temp| mean','temp| max',
                          'wspd| min','wspd| mean','wspd| max',
                          'pressure| min','pressure| mean','pressure| max',
                          'rh| min','rh| mean','rh| max'
                         ]].corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

sn.heatmap(corr, mask=mask,  vmax=.3, center=0,annot=True,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```
