```python
city_keys={'Tasmania':'YMHB:9:AU',
           'New York':'KLGA:9:US',
           'Virginia':'KLYH:9:US',
           'Quebec':'CYQB:9:CA',
           'New South Wales':'YCBA:9:AU',
           'Henan':'ZHCC:9:CN',
           'Bermuda':'TXKF:9:BM',
           'Reunion':'FMEE:9:RE',
           'Lombardia':'LIML:9:IT',
           'Maharashtra':'VABB:9:IN'}
```

```python
for keys in city_keys.items():
    df=pd.DataFrame()
    for month in pd.date_range(start='20200101',end='20200601',freq='M'):
        start,end=month.replace(day=1).strftime('%Y%m%d'),month.strftime('%Y%m%d')
        url="https://api.weather.com/v1/location/"+keys[1]+"/observations/historical.json?apiKey=6532d6454b8aa370768e63d6ba5a832e&units=e&startDate="+start+"&endDate="+end        
        res = requests.get(url)
        json_text=json.loads(res.text)
        df=pd.concat([df,pd.DataFrame(json_text['observations'])])
    df.to_csv('../../../datasets/created/weather/Weather '+keys[0]+'.csv',index=False) 
    print(keys[0])

    
```
